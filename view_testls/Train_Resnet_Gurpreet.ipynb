```python
from torchvision import datasets,models,transforms
import torch.nn as nn
from nndev import model_pip
 

for j in xrange(0,10):
    k=j+1
    setnumber="SET"+str(k)
    Model_name="ResNetModel_Pretrained_7_views_bs_100_e_10_28062017_0640.pth.tar"
    datapath="/data/Gurpreet/RUNS/VC_10/"+setnumber+"/dataset/"
    savepath="/data/Gurpreet/RUNS/VC_10/"+setnumber+"/"+Model_name
    res = models.resnet18(pretrained=True)
    res.fc = nn.Linear(res.fc.in_features,7)
    obj = model_pip(model_in=res,scale=True,batch_size=100,use_gpu=True,gpu=1,data_path=datapath)
    model = obj.train_model(epochs=10)
    obj.store_model(f_name=savepath)
    mdir="/data/Gurpreet/RUNS/VC_10/"+str(setnumber)+"/"
      
    obj.test(model_dir=mdir,model_name=Model_name,save_miscl = True,test_on=True,n='None',random_crop=False,folder='misc')
    #obj.test(model_dir=mdir,test_on=True,model_name="ResNetModel_Pretrained.pth.tar",save_miscl = True)
```
