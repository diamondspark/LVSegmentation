```python
from torchvision import datasets,models,transforms
import torch.nn as nn
from nndev import model_pip
#CUDA_LAUNCH_BLOCKING=1 

for j in xrange(0,10):
    k=j+1
    setnumber="SET"+str(k)
    Model_name="ResNetModel_Pretrained_3_views_bs_30_e_20_28062017_0705.pth.tar"
    datapath="/data/Gurpreet/RUNS/VC_11/"+setnumber+"/dataset/"
    savepath="/data/Gurpreet/RUNS/VC_11/"+setnumber+"/ResNetModel.pth.tar"
    res = models.resnet18(pretrained=True)
    res.fc = nn.Linear(res.fc.in_features,3)
    obj = model_pip(model_in=res,scale=True,batch_size=30,use_gpu=True,gpu=2,data_path=datapath)
    model = obj.train_model(epochs=20)
    obj.store_model(f_name=savepath)
    mdir="/data/Gurpreet/RUNS/VC_10/"+str(setnumber)+"/"
    obj.test(model_dir=mdir,model_name=Model_name,save_miscl = True,test_on=True,n='None',random_crop=False,folder='misc')
    del (obj)
```
